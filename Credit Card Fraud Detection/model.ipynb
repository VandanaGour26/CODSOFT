```python
training_data.drop(columns=['Unnamed: 0','first','last','job','gender','dob','cc_num','merchant','trans_num'
                            ,'street','city','state','zip'],inplace=True)
# remove card no (cc_num) becoz model card ke no se sikh lega or wrong prediction 
```

```python
training_data['trans_date_trans_time'] = pd.to_datetime(training_data['trans_date_trans_time'])
training_data['trans_date'] = training_data['trans_date_trans_time'].dt.date
training_data['trans_time'] = training_data['trans_date_trans_time'].dt.time
```

```python
training_data.drop(columns=['trans_date_trans_time'],inplace=True)
```

```python
training_data.head(50)
```

```python
#handle_unknown='ignore' ye ignore karega if any other category comes in test data but was not in training data
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
training_data['category'] = encoder.fit_transform(training_data[['category']])
training_data
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
training_data['amt'] =  scaler.fit_transform(training_data[['amt']])
training_data['city_pop'] =  scaler.fit_transform(training_data[['city_pop']])
training_data
```

```python
training_data['is_fraud'].value_counts()
```

```python
y = training_data['is_fraud']
x = training_data.drop(columns='is_fraud')
```

```python
x
```

```python
y
```

```python

x.drop(columns=['trans_date','trans_time'], inplace=True)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
logistic = LogisticRegression(max_iter=200)
decision = DecisionTreeClassifier(random_state=42)
```

```python
logistic_scores = cross_val_score(
    logistic,
    x,
    y,
    cv=skf,
    scoring='roc_auc'
)

print("Logistic Regression CV ROC-AUC:", logistic_scores)
print("Mean:", logistic_scores.mean())
```

```python
decision_scores = cross_val_score(
    decision,
    x,
    y,
    cv=skf,
    scoring='roc_auc'
)

print("Decision Tree CV ROC-AUC:", decision_scores)
print("Mean:", decision_scores.mean())
```
