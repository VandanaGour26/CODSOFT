```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df
```

```python
df['Geography'] = encoder.fit_transform(df['Geography'])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['CreditScore'] = scaler.fit_transform(df[['CreditScore']]).ravel()
df['Age'] = scaler.fit_transform(df[['Age']]).ravel()
df['Balance'] = scaler.fit_transform(df[['Balance']]).ravel()
df['EstimatedSalary'] = scaler.fit_transform(df[['EstimatedSalary']]).ravel()
```

```python


df
```

```python
df['Exited'].value_counts()
#imbalance data
```

```python
x =df.drop(columns=['Exited']) 
y = df['Exited']
x
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
gb = GradientBoostingClassifier(random_state=42)
logistic = LogisticRegression(max_iter=200)
random = RandomForestClassifier(random_state=42, class_weight='balanced')
```

```python
logistic_scores = cross_val_score(
    logistic,
    train_x,
    train_y,
    cv=skf,
    scoring='roc_auc'
)

print("Logistic Regression CV ROC-AUC:", logistic_scores)
print("Mean:", logistic_scores.mean())
```

```python
random_scores = cross_val_score(
    random,
    train_x,
    train_y,
    cv=skf,
    scoring='roc_auc'
)

print("Random Forest CV ROC-AUC:", random_scores)
print("Mean:", random_scores.mean())
```

```python
gb_scores = cross_val_score(
    gb,
    train_x,
    train_y,
    cv=skf,
    scoring='roc_auc'
)

print("Logistic Regression CV ROC-AUC:", gb_scores)
print("Mean:", gb_scores.mean())
```

```python
logistic.fit(x, y)
```

```python
random.fit(x, y)
```

```python
gb.fit(x, y)
```

```python
logistic_predict=logistic.predict(test_x)
random_predict=random.predict(test_x)
gb_predict=gb.predict(test_x)
```

```python
from sklearn.metrics import roc_auc_score

print("logistic score:",roc_auc_score(test_y, logistic_predict))
print("random forest score:",roc_auc_score(test_y, random_predict))
print("gb score:",roc_auc_score(test_y, gb_predict))
```
